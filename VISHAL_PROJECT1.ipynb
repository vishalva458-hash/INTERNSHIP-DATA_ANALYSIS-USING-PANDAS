### TOTAL REVENUE

```python
df["TotalAmount"] = df["Total Amount After Discount"]
total_revenue = df["TotalAmount"].sum()
print("Total Revenue: ₹", total_revenue)
```

### REVENUE BY CATEGORY

```python
revenue_by_category = df.groupby("Category")["TotalAmount"].sum()
print("\nRevenue by Category:\n", revenue_by_category)
# Calculate revenue by category
revenue_by_category = df.groupby("Category")["TotalAmount"].sum()
# Plot
import matplotlib.cm as cm
plt.figure(figsize=(10, 4))
plt.bar(revenue_by_category.index, revenue_by_category.values)
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=0)
colors = cm.Set3(range(len(revenue_by_category)))
plt.bar(revenue_by_category.index, revenue_by_category.values, color=colors)
plt.tight_layout()
plt.show()
```

### REVENUE BY CITY

```python
revenue_by_city = df.groupby("City")["TotalAmount"].sum()
print("\nRevenue by City:\n", revenue_by_city)
# plot
import matplotlib.pyplot as plt
# Calculate revenue by city
revenue_by_city = df.groupby("City")["TotalAmount"].sum()
# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    revenue_by_city.values,
    labels=revenue_by_city.index,
    autopct="%1.1f%%",
    startangle=140
)
plt.title("Revenue Share by City")
plt.tight_layout()
plt.show()
```

### QUANTITY PURCHASED

```python
# Calculate total quantity purchased per product
product_quantity = df.groupby("Product")["Quantity"].sum()
# Plot 
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(product_quantity.index, product_quantity.values, marker='o')
max_product = product_quantity.idxmax()
min_product = product_quantity.idxmin()
plt.scatter(max_product, product_quantity[max_product], color='red', s=100)    # highlight max
plt.scatter(min_product, product_quantity[min_product], color='red', s=100)  # highlight min
plt.title("Quantity Purchased per Product")
plt.xlabel("Product")
plt.ylabel("Total Quantity Purchased")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

import seaborn as sns
product_quantity = df.groupby("Product")["Quantity"].sum().reset_index()
plt.figure(figsize=(2, 3))
sns.boxplot(data=product_quantity, y="Quantity")
sns.swarmplot(data=product_quantity, y="Quantity", color="red", size=4)
plt.title("Box + Swarm Plot of Quantity Purchased")
plt.ylabel("Total Quantity Purchased")
plt.tight_layout()
plt.show()
```



### MOST PURCHASED PRODUCT

```python
most_purchased_product = df.groupby("Product")["Quantity"].sum().idxmax()
print("\nMost Purchased Product:", most_purchased_product)
```

### LEAST PURCHASED PRODUCT¶

```python
least_purchased_product = df.groupby("Product")["Quantity"].sum().idxmin()
print("\nLeast Purchased Product:", least_purchased_product)
```

### CUSTOMER WISE REVENUE

```python
customer_revenue = df.groupby("Customer Name")["TotalAmount"].sum()
print("\nCustomer-wise Revenue:\n", customer_revenue)
# Calculate customer-wise revenue
customer_revenue = df.groupby("Customer Name")["TotalAmount"].sum()
# Plot
plt.figure(figsize=(9,5 ))
plt.bar(customer_revenue.index, customer_revenue.values)
plt.title("Customer-wise Revenue")
plt.xlabel("Customer Name")
plt.ylabel("Total Revenue")
plt.xticks(rotation=90)   # Rotate labels for readability
colors = cm.tab20(range(len(customer_revenue))) 
plt.bar(customer_revenue.index, customer_revenue.values, color=colors)  
plt.tight_layout()
plt.show()
```

### DAILY REVENUE TREND

```python
df["Delivery Date"] = pd.to_datetime(df["Delivery Date"])
daily_revenue = df.groupby(df["Delivery Date"].dt.date)["TotalAmount"].sum()
print("\nDaily Revenue Trend:\n", daily_revenue)
# Convert delivery date and calculate daily revenue
df["Delivery Date"] = pd.to_datetime(df["Delivery Date"])
daily_revenue = df.groupby(df["Delivery Date"].dt.date)["TotalAmount"].sum()
# Plot line graph
plt.figure(figsize=(11, 4))
plt.plot(daily_revenue.index, daily_revenue.values, marker='o')
plt.title("Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Total Revenue")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## ADVANCED INSIGHTS

### AVERAGE ORDER VALUE 

```python
aov = df["TotalAmount"].mean()
print("\nAverage Order Value (AOV): ₹", aov)
```

### PAYMENT PATH

```python
payment_mode_count = df["Payment Mode"].value_counts()
print("\nPayment Mode Usage Count:\n", payment_mode_count)
# Count payment mode usage
payment_mode_count = df["Payment Mode"].value_counts()
# Plot pie chart
plt.figure(figsize=(5, 4))
plt.pie(
    payment_mode_count.values,
    labels=payment_mode_count.index,
    autopct="%1.1f%%",
    startangle=140
)
plt.title("Payment Mode Usage Distribution")
plt.tight_layout()
plt.show()
```

### CATEGORY WISE AVERAGE PRICE

```python
category_avg_price = df.groupby("Category")["Price"].mean()
print("\nCategory-wise Average Price:\n", category_avg_price)
# Calculate average price by category
category_avg_price = df.groupby("Category")["Price"].mean()
# Plot bar chart
colors = cm.tab20(range(len(category_avg_price)))   
plt.figure(figsize=(10, 4))
plt.bar(category_avg_price.index, category_avg_price.values, color=colors)   
plt.title("Category-wise Average Price")
plt.xlabel("Category")
plt.ylabel("Average Price")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### TOPCITY FOR FASHION 

```python
fashion_city = (
    df[df["Category"] == "Fashion"]
    .groupby("City")["TotalAmount"]
    .sum()
    .idxmax()
)
print("\nTop City for Fashion Category:", fashion_city)
# Filter Fashion category and calculate revenue by city
fashion_city_revenue = (
    df[df["Category"] == "Fashion"]
    .groupby("City")["TotalAmount"]
    .sum()
)
# Plot bar chart
colors = cm.tab20(range(len(fashion_city_revenue)))   
plt.figure(figsize=(8, 4))
plt.bar(fashion_city_revenue.index, fashion_city_revenue.values, color=colors)   
plt.title("Revenue by City for Fashion Category")
plt.xlabel("City")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
